```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
CKPT_FILEPATH = 'checkpoint/baseline - RNN/sample/0037.npy'
GRAPH_FILEPATH = 'checkpoint/baseline - RNN/sample-0037.png'
```

```python
data = np.load(CKPT_FILEPATH)
```

```python
data.shape
```

```python
yaw_gt, pitch_gt, roll_gt = data[0], data[1], data[2]
yaw, pitch, roll = data[3], data[4], data[5]
```

```python
title = 'RNN'
s = 1500
t = 2100

plt.figure(figsize=(16, 3))

plt.subplot(1, 3, 1)
plt.plot(yaw_gt[s:t])
plt.plot(yaw[s:t])
plt.legend(['yaw_gt', 'yaw'])
plt.title(title + ' - yaw')
plt.ylim(-1, 1)

plt.subplot(1, 3, 2)
plt.plot(pitch_gt[s:t])
plt.plot(pitch[s:t])
plt.legend(['pitch_gt', 'pitch'])
plt.title(title + ' - pitch')
plt.ylim(-1, 1)

plt.subplot(1, 3, 3)
plt.plot(roll_gt[s:t])
plt.plot(roll[s:t])
plt.legend(['roll_gt', 'roll'])
plt.title(title + ' - roll')
plt.ylim(-1, 1)

plt.tight_layout()
plt.savefig(GRAPH_FILEPATH)
plt.show()
```
