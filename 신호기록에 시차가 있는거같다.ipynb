```python
import numpy as np
import matplotlib.pyplot as plt
import os
from pathlib import Path
import pandas as pd
```

```python
def plot(*data):
    dd = []
    ls = []
    for d in data:
        if isinstance(d, str):
            ls.append(d)
        else:
            dd.append(d)
    
    plt.figure(figsize=(16, 4))
    for d in dd:
        plt.plot(d)
    plt.legend(ls)
    plt.show()
```

```python
def norm(data):
    return (data - data.min()) / (data.max() - data.min())
```

```python
DATA_DIR = Path('org')
```

```python
CSV_FILES = [pd.read_csv(f) for f in sorted(list(DATA_DIR.glob('*.csv')))]
```
