```python
import numpy as np
import pandas as pd
import imageio
import h5py
import pickle

from tqdm import tqdm
from xqdm import xqdm
from pathlib import Path
import matplotlib.pyplot as plt
import re

import keras.layers as L
from keras.models import Model
from keras.utils import plot_model
```

```python
I = L.Input(shape=(18,))
x = L.Dense(32, activation='relu')(I)
x = L.Dense(128, activation='relu')(x)
x = L.Dropout(0.2)(x)
x = L.Dense(256, activation='relu')(x)
x = L.Dropout(0.2)(x)
x = L.Dense(1, activation='linear')(x)

model = Model(inputs=I, outputs=x)
```

```python
model.summary()
```

```python
model.compile('adam', loss='mean_squared_error')
```

```python
with open('data/mlp18_pitch.pkl', 'rb') as f:
    X, Y = pickle.load(f)

#X = np.stack(X, axis=0)
#Y = np.stack(Y, axis=0)

print(X.shape, Y.shape)
```

```python
model.fit(X, Y, batch_size=64, epochs=100, verbose=1, validation_split=0.2, shuffle=False)
```
