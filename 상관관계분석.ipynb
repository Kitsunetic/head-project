```python
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt
from math import pi
import pandas as pd
import seaborn as sns
```

```python
for u in range(1, 8):
    for s in range(1, 4):
        for t in range(2):
            csv = pd.read_csv(f'data/1108/user{u}_scene{s}_{t}.csv')

            plt.figure(figsize=(6, 6))
            sns.heatmap(csv.corr(), annot=True, fmt='.2f', linewidths=.5, cmap='Blues')
            plt.title(f'data/1108/user{u}_scene{s}_{t}.csv')
            plt.tight_layout()
            plt.savefig(f'img/heatmap/user{u}_scene{s}_{t}.png')
            plt.show()
```
