# 1116 Scene1 RNN

```python
data = np.load('checkpoint/Scene1-MLP/sample/0017.npy')
```

```python
target = data[:3]
output = data[3:]
```

```python
target.shape
```

```python
s = 50000
t = 55000
X = np.linspace(s / 60, t / 60, t - s)
plt.plot(X, target[0][s:t])
plt.plot(X, output[0][s:t])
plt.legend(['yaw; target', 'yaw; prediction'])
plt.ylabel('degree')
plt.xlabel('time (s)')
plt.tight_layout()
#plt.savefig('img/1116-yaw-LSTM1, 많은 움직임, 잘됨.png')
plt.show()
```
