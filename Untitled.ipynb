```python
import torch
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
csv = pd.read_csv('data/1116/user1_scene1_0.csv')
```

```python
plt.plot(csv.timestamp, csv.input_orientation_roll)
plt.legend(['roll'])
plt.xlabel('time (sec)')
plt.ylabel('degree')
plt.title('Roll in Scene1 User1')
plt.tight_layout()
plt.savefig('img/scene1에서의 roll.png')
plt.show()
```

```python
plt.plot(csv.timestamp, csv.input_orientation_yaw)
plt.legend(['yaw'])
plt.xlabel('time (sec)')
plt.ylabel('degree')
plt.title('Yaw in Scene1 User1')
plt.tight_layout()
plt.savefig('img/scene1에서의 yaw.png')
plt.show()
```
