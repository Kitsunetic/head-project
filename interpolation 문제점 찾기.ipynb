```python
import torch
import torch.nn as nn
import torchaudio.transforms as AT
import h5py
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csv = pd.read_csv('data/interpolation/interpolation_00.csv')
```

```python
yaw = csv['input_orientation_yaw']
pitch = csv['input_orientation_pitch']
roll = csv['input_orientation_roll']
```

```python
dyaw = yaw[1:] - yaw[:-1]
dpitch = pitch[1:] - pitch[:-1]
droll = roll[1:] - roll[:-1]
```

```python
for i in range(22):
    csv = pd.read_csv(f'data/interpolation/interpolation_{i:02d}.csv')
    yaw = csv['input_orientation_yaw'].to_numpy()
    pitch = csv['input_orientation_pitch'].to_numpy()
    roll = csv['input_orientation_roll'].to_numpy()
    dyaw = yaw[1:] - yaw[:-1]
    dpitch = pitch[1:] - pitch[:-1]
    droll = roll[1:] - roll[:-1]

    plt.figure(figsize=(16, 3))
    plt.plot(dyaw)
    plt.title(f'{i:02d} dyaw')
    plt.ylim(-0.5, 0.5)
    plt.tight_layout()
    plt.show()
    plt.figure(figsize=(16, 3))
    plt.plot(dpitch)
    plt.title(f'{i:02d} dpitch')
    plt.ylim(-0.5, 0.5)
    plt.tight_layout()
    plt.show()
    plt.figure(figsize=(16, 3))
    plt.plot(droll)
    plt.title(f'{i:02d} droll')
    plt.ylim(-0.5, 0.5)
    plt.tight_layout()
    plt.show()
```
