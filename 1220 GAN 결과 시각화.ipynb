```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.load('results4_gan/0002/data-epoch158.npz')
```

```python
X = data['X']
Y = data['Y']
P = data['P']
```

```python
s = 200
t = 1200
T = np.linspace(s / 600, t / 600, t-s)

plt.plot(T, X[s:t, 0])
plt.plot(T, Y[s:t, 0])
plt.plot(T, P[s:t, 0])
plt.legend(['input', 'real', 'predict'])
plt.tight_layout()
plt.show()
```
