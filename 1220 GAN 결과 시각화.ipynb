```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.load('results4_gan/0004-ForScore/data-epoch10.npz')

X = data['X'].flatten()
Y = data['Y'].flatten()
P = data['P'].flatten()
```

```python
X.shape, Y.shape, P.shape
```

```python
s = 9000
t = 10000
T = np.linspace(s / 600, t / 600, t-s)

plt.figure(figsize=(12, 3))
plt.plot(T, X[s:t, 0])
plt.plot(T, Y[s:t, 0])
plt.plot(T, P[s:t, 0])
plt.legend(['input', 'real', 'predict'])
plt.tight_layout()
plt.show()
```
